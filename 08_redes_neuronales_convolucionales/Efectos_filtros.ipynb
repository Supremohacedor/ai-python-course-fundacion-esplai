```python
# ---------------------------------------------
# Ejemplo didctico: Visualizacin de filtros en imgenes (CNN Basics)
# ---------------------------------------------

# 1. Importar libreras necesarias
import numpy as np
import matplotlib.pyplot as plt
from skimage import data, color
from scipy.ndimage import convolve

# 2. Cargar imagen de ejemplo y convertir a escala de grises
image = color.rgb2gray(data.astronaut())  # Puedes cambiar 'astronaut' por otra imagen de skimage

# 3. Definir algunos filtros clsicos de visin computacional
filtros = {
    "Original": np.eye(3),
    "Bordes Verticales (Sobel)": np.array([[-1, 0, 1],
                                           [-2, 0, 2],
                                           [-1, 0, 1]]),
    "Bordes Horizontales (Sobel)": np.array([[1, 2, 1],
                                             [0, 0, 0],
                                             [-1, -2, -1]]),
    "Desenfoque (Blur)": np.ones((9, 9)) / 9, # Cuanto ms grande, ms se desenfoca
    "Realce (Sharpen)": np.array([[0, -1, 0],
                                  [-1, 5, -1],
                                  [0, -1, 0]]),
    "Relieve (Emboss)": np.array([[-2, -1, 0],
                                  [-1, 1, 1],
                                  [0, 1, 2]])
}

# 4. Aplicar cada filtro y mostrar resultados
fig, axes = plt.subplots(2, 3, figsize=(14, 9))
for ax, (nombre, filtro) in zip(axes.flat, filtros.items()):
    if nombre == "Original":
        imagen_filtrada = image
    else:
        imagen_filtrada = convolve(image, filtro)
    ax.imshow(imagen_filtrada, cmap='gray')
    ax.set_title(nombre)
    ax.axis('off')

plt.tight_layout()
plt.show()
```
