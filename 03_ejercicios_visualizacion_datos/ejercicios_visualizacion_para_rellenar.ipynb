## 1. Barplot: Distribuci√≥n de calificaciones

**Dataset:** [Students Performance in Exams (Kaggle)](https://www.kaggle.com/datasets/spscientist/students-performance-in-exams)

**Pasos:**
1. Descarga el archivo CSV y c√°rgalo con pandas.
2. Realiza un barplot que muestre **el n√∫mero de estudiantes por grupo √©tnico** (`race/ethnicity`).

**Preguntas:**
- ¬øQu√© grupo es el m√°s numeroso?
- ¬øHay grupos con representaci√≥n especialmente baja?


```python
import seaborn as sns
import pandas as pd
import matplotlib.pyplot as plt

students_performance = pd.read_csv('StudentsPerformance.csv')

plt.figure(figsize=(7,5))
sns.countplot(x='race/ethnicity', data=students_performance)
plt.title('N√∫mero de estudiantes por grupo √©tnico')
plt.xlabel('Grupo √©tnico')
plt.ylabel('N√∫mero de estudiantes')
plt.show()

'''
**Preguntas:**
- ¬øQu√© grupo es el m√°s numeroso?
El grupo m√°s numeroso es el grupo C.
- ¬øHay grupos con representaci√≥n especialmente baja?
Los grupos con representaci√≥n especialmente baja son los grupos A y E.
'''
```

## 8. Line Plot: Series de tiempo de temperaturas

**Dataset:** Cualquier dataset de temperaturas globales (ej: Kaggle, NOAA).

**Pasos:**
1. Descarga y carga el dataset con pandas.
2. Calcula la **temperatura media anual**.
3. Haz un line plot mostrando la evoluci√≥n de la temperatura a lo largo del tiempo.

**Preguntas:**
- ¬øSe observa una tendencia clara?
- ¬øHay a√±os con valores an√≥malos?


```python
global_temperature = pd.read_csv('temperature_sample.csv')

#print(global_temperature.columns)

#Me quedo solo con el a√±o y lo convierto a int
global_temperature['Year'] = global_temperature['Year'].str[:4]
global_temperature['Year'] = global_temperature['Year'].astype(int)

#Saco la media de temperatura por a√±o
avg_temp = global_temperature.groupby('Year')['Mean'].mean()

plt.figure(figsize=(20,5))
plt.plot(avg_temp.index, avg_temp.values, marker='o')
plt.title('Curva de temperatura media anual a lo largo de los a√±os')
plt.xlabel('A√±o')
plt.ylabel('Temperaura media anual')
plt.grid(True)
plt.show()

'''
**Preguntas:**
- ¬øSe observa una tendencia clara?
S√≠, la temperatura media anual ha ido aumentando.
- ¬øHay a√±os con valores an√≥malos?
S√≠, los a√±os con valores an√≥malos se encuentran cerca de 1875.
'''
```
